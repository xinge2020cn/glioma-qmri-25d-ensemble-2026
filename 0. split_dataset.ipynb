```python
import os  
import pandas as pd  
  

source_dir = ''  
target_dir = ''  
  

if not os.path.exists(target_dir):  
    os.makedirs(target_dir)  
  

for filename in os.listdir(source_dir):  
    if filename.endswith('.csv'):  

        file_path = os.path.join(source_dir, filename)  
        df = pd.read_csv(file_path)  
  

        df_train = df[df['group'] == 'train'].sort_values('ID')  
        df_val = df[df['group'] == 'val'].sort_values('ID')  
        df_test = df[df['group'] == 'test'].sort_values('ID')  
  

        if len(df_test) >= 100:  
            df_val_new = pd.concat([df_val, df_test.head(100).assign(group='val')])  
            df_test_new = df_test.iloc[100:]  
        else:  
            df_val_new = df_val  
            df_test_new = df_test  
  

        if not df_test_new.empty:  
            df_test_new = df_test_new.assign(group='test')  
  

        df_sorted = pd.concat([df_train, df_val_new, df_test_new], ignore_index=True)  
  

        target_file_path = os.path.join(target_dir, filename)  
        df_sorted.to_csv(target_file_path, index=False)
  
print("", target_dir)
```
